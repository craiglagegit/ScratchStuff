```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import lsst.sims.maf.db as db
import lsst.sims.maf.metrics as metrics
import lsst.sims.maf.slicers as slicers
import lsst.sims.maf.stackers as stackers
import lsst.sims.maf.metricBundles as metricBundles
from lsst.sims.maf.plots import NeoDetectPlotter
import lsst.sims.maf.plots as plotters
```

```python
# Set up the database connection
opsdb = db.OpsimDatabase('sqlite:///enigma_1189_sqlite.db')
outDir = 'NeoEclip'
resultsDb = db.ResultsDb(outDir=outDir)
configSum,config =opsdb.fetchConfig()
```

```python
propIDs=[]
propNames=[]
for key in configSum['Proposals']['keyorder']:
    propNames.append(key)
    propIDs.append(configSum['Proposals'][key]['propID'])
```

```python
metric = metrics.CountMetric('expMJD')
slicer = slicers.HealpixSlicer(nside=64, latCol='eclipLat',lonCol='eclipLon')
stacker = stackers.EclipticStacker(subtractSunLon=False)
sqls = ['propID = %s' % propID for propID in propIDs]

for sql,propName in zip(sqls,propNames):
    plotDict = {'title':propName}

    bundle = metricBundles.MetricBundle(metric, slicer,sql,stackerList=[stacker], plotDict=plotDict)
    bgroup = metricBundles.MetricBundleGroup({0:bundle}, opsdb, outDir=outDir, resultsDb=resultsDb)
    bgroup.runAll()
    bgroup.plotAll(closefigs=False)
```

Now the same thing, but in the rotating frame with the sun at longitude 0

```python
metric = metrics.CountMetric('expMJD')
slicer = slicers.HealpixSlicer(nside=64, latCol='eclipLat',lonCol='eclipLon')
stacker = stackers.EclipticStacker(subtractSunLon=True)
sqls = ['propID = %s' % propID for propID in propIDs]

for sql,propName in zip(sqls,propNames):
    plotDict = {'rot':(180,0,0), 'title':propName}

    bundle = metricBundles.MetricBundle(metric, slicer,sql,stackerList=[stacker], plotDict=plotDict)
    bgroup = metricBundles.MetricBundleGroup({0:bundle}, opsdb, outDir=outDir, resultsDb=resultsDb)
    bgroup.runAll()
    bgroup.plotAll(closefigs=False)
```
